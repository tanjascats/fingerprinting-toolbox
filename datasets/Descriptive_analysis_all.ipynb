## German Credit 

```python
# fetch dataset 
statlog_german_credit_data = fetch_ucirepo(id=144) 
  
# data (as pandas dataframes) 
X = statlog_german_credit_data.data.features 
y = statlog_german_credit_data.data.targets 
```

```python
# Get the list of attribute names
#attribute_names = X.columns
attribute_names = ['checking_account', 'duration', 'credit_hist', 'purpose', 'credit_amount',
                   'savings', 'employment_since', 'installment_rate', 'sex_status', 'debtors',
                   'residence_since', 'property', 'age', 'installment_other', 'housing',
                   'existing_credits','job','liable_people','tel','foreign']
X.columns = attribute_names

# Calculate the number of rows and columns for the grid
num_attributes = len(attribute_names)
num_cols = 3
num_rows = (num_attributes + num_cols - 1) // num_cols  # Calculate the number of rows

# Create a new figure and set the size
fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 5 * num_rows))

# Loop through the attributes and create histograms or bar plots
for i, attribute in enumerate(attribute_names):
    row = i // num_cols
    col = i % num_cols

    if X[attribute].dtype == 'object':
        # For categorical data, create a bar plot
        X[attribute].value_counts().plot(kind='bar', ax=axes[row, col])
    else:
        # For numerical data, create a histogram
        X[attribute].plot(kind='hist', ax=axes[row, col], bins=10)
    axes[row, col].set_title(attribute)
    axes[row, col].grid(axis='y')

# Remove any empty subplots if the number of attributes is not a multiple of 3
for i in range(num_attributes, num_rows * num_cols):
    fig.delaxes(axes.flatten()[i])

# Adjust layout and display the plots
plt.tight_layout()
plt.savefig('plots/german_credit.svg')
plt.savefig('plots/german_credit.png')
plt.show()
```

## Adult

```python
# fetch dataset 
adult = fetch_ucirepo(id=2) 
  
# data (as pandas dataframes) 
X = adult.data.features 
y = adult.data.targets 
```

```python
# Get the list of attribute names
attribute_names = X.drop(['fnlwgt', 'education-num'], axis=1).columns
# X.columns = attribute_names

# Calculate the number of rows and columns for the grid
num_attributes = len(attribute_names)
num_cols = 3
num_rows = (num_attributes + num_cols - 1) // num_cols  # Calculate the number of rows

# Create a new figure and set the size
fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 5 * num_rows))

# Loop through the attributes and create histograms or bar plots
for i, attribute in enumerate(attribute_names):
    row = i // num_cols
    col = i % num_cols
    if X[attribute].dtype == 'object':
        # For categorical data, create a bar plot
        if attribute == 'native-country':
            native_country = X['native-country'].value_counts()[:9]
            native_country['other'] = X['native-country'].value_counts()[9:].sum()
            native_country.plot(kind='bar', ax=axes[row, col])
        else:
            X[attribute].value_counts().plot(kind='bar', ax=axes[row, col])
    else:
        # For numerical data, create a histogram
        X[attribute].plot(kind='hist', ax=axes[row, col], bins=10)
    axes[row, col].set_title(attribute)
    axes[row, col].grid(axis='y')

# Remove any empty subplots if the number of attributes is not a multiple of 3
for i in range(num_attributes, num_rows * num_cols):
    fig.delaxes(axes.flatten()[i])

# Adjust layout and display the plots
plt.tight_layout()
plt.savefig('plots/adult_cut.svg')
plt.savefig('plots/adult_cut.png')
plt.show()
```
